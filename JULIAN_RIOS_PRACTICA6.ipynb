<h1>Hipótesis nula y alternativa para la predicción del valor de los fondos de inversión en Colombia</h1>
<h2>Hipótesis nula (H0):</h2>

No existe una relación significativa entre las variables independientes (categoría del fondo, riesgo, rating, etc.) y la rentabilidad del fondo.

<h2>Hipótesis alternativa (H1)</h2>:

Existe una relación significativa entre las variables independientes y la rentabilidad del fondo.

<h2>Explicación:</h2>

La hipótesis nula establece que no hay relación entre las variables independientes y la variable dependiente (rentabilidad del fondo). Esto significa que la rentabilidad del fondo es aleatoria y no está influenciada por las características del fondo.
La hipótesis alternativa establece que sí hay una relación entre las variables independientes y la variable dependiente. Esto significa que la rentabilidad del fondo está influenciada por las características del fondo.

<h2>Mutua exclusión:</h2>

Las hipótesis nula y alternativa son mutuamente excluyentes. No pueden ser ambas verdaderas al mismo tiempo. Si la hipótesis nula es verdadera, entonces la hipótesis alternativa es falsa. Y viceversa.

<h2>Colectiva exhaustividad:</h2>

La hipótesis nula y la alternativa se pueden verificar mediante un test estadístico, como el test de Wald o el test de razón de verosimilitud.
El test estadístico nos dará un valor p, que es la probabilidad de obtener un resultado tan extremo como el que se ha observado si la hipótesis nula fuera verdadera.
Si el valor p es menor que el nivel de significancia elegido (generalmente 0.05), entonces se rechaza la hipótesis nula y se acepta la hipótesis alternativa.

<h2>Ejemplo de aplicación:</h2>

H0: No existe una diferencia significativa en la rentabilidad entre los fondos de renta fija y los fondos de renta variable.
H1: Los fondos de renta variable tienen una mayor rentabilidad que los fondos de renta fija.

En este caso, la variable independiente es el tipo de fondo (renta fija o renta variable) y la variable dependiente es la rentabilidad del fondo.

# Regresión logistica

```python
# Selección de variables
X = df[['Volatilidad']]  # Variable continua
#  variable binaria a partir de la rentabilidad diaria
y = (df['Rentabilidad diaria'] > 0).astype(int)

# División de los datos en conjuntos de entrenamiento y prueba
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Creación del modelo de regresión logística
model = LogisticRegression()

# Ajuste del modelo a los datos de entrenamiento
model.fit(X_train, y_train)

# Obtención de los coeficientes de regresión
coeficientes = model.coef_
print("Coeficientes de regresión:", coeficientes)

# Obtención del término de intercepción
intercepcion = model.intercept_
print("Término de intercepción:", intercepcion)

# Realización de predicciones en el conjunto de prueba
y_pred = model.predict(X_test)

# Cálculo de la precisión del modelo
precision = accuracy_score(y_test, y_pred)
print("Precisión del modelo:", precision)

# Cálculo de la matriz de confusión
matriz_confusion = confusion_matrix(y_test, y_pred)
print("Matriz de confusión:")
print(matriz_confusion)
```

<b>Coeficiente de regresión:</b> El coeficiente de regresión es 0.09243351. Esto significa que, para un incremento unitario en la variable de entrada (en este caso, la volatilidad), se espera un aumento de aproximadamente 0.0924 en el logaritmo de la odds de pertenecer a la clase positiva (rentabilidad diaria mayor que cero), manteniendo todas las demás variables constantes. Es importante tener en cuenta que este valor es positivo, lo que sugiere que un aumento en la volatilidad está asociado con una mayor probabilidad de obtener una rentabilidad diaria positiva.

<b>Término de intercepción:</b> El término de intercepción es -0.01722082. Esto representa la probabilidad de pertenecer a la clase positiva cuando todas las características son iguales a cero (en este caso, volatilidad igual a cero). Dado que este valor es negativo, significa que cuando la volatilidad es cero, la probabilidad de obtener una rentabilidad diaria positiva es menor.

<b>Precisión del modelo:</b> La precisión del modelo es aproximadamente 0.4989 o 49.89%. Esto indica que el modelo clasifica correctamente alrededor del 49.89% de las muestras en el conjunto de prueba.

<b>Matriz de confusión:</b>

<b>El elemento en la posición (1,1)</b> indica el número de verdaderos positivos (VP), es decir, la cantidad de muestras clasificadas correctamente como positivas.

```python
# Dividir los datos en características (X) y etiquetas (y)
X = df_final[['Volatilidad', 'Benchmark', 'Sharpe Ratio']]  # Características
y = (df_final['Rentabilidad diaria'] > 0).astype(int)  # Etiquetas (0 o 1)

# Dividir los datos en conjuntos de entrenamiento y prueba (70% entrenamiento, 30% prueba)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Crear el modelo de regresión logística
model = LogisticRegression()

# Ajustar el modelo a los datos de entrenamiento
model.fit(X_train, y_train)

# Realizar predicciones en el conjunto de prueba
y_pred = model.predict(X_test)

# Calcular la precisión del modelo
precision = accuracy_score(y_test, y_pred)
print("Precisión del modelo:", precision)

# Calcular la matriz de confusión del modelo
matriz_confusion = confusion_matrix(y_test, y_pred)
print("Matriz de confusión:")
print(matriz_confusion)
```

```python
# Guardar los nombres de las columnas antes de escalar los datos
cols = X_train.columns

# Escalar los datos de entrenamiento y prueba
scaler = MinMaxScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)

# Convertir las matrices escaladas nuevamente a DataFrames con los nombres de las columnas originales
X_train = pd.DataFrame(X_train, columns=cols)
X_test = pd.DataFrame(X_test, columns=cols)

# Instanciar el modelo de regresión logística
logreg = LogisticRegression(solver='liblinear', random_state=0)

# Ajustar el modelo a los datos de entrenamiento
logreg.fit(X_train, y_train)

# Predecir las etiquetas del conjunto de prueba
y_pred_test = logreg.predict(X_test)

# Imprimir las etiquetas predichas para el conjunto de prueba
print(y_pred_test)

# Predecir las probabilidades de las clases para el conjunto de prueba
probs_class_0 = logreg.predict_proba(X_test)[:,0]
probs_class_1 = logreg.predict_proba(X_test)[:,1]

# Imprimir las probabilidades de pertenecer a la clase 0 y clase 1 para el conjunto de prueba
print(probs_class_0)
print(probs_class_1)

# Predecir las etiquetas del conjunto de entrenamiento
y_pred_train = logreg.predict(X_train)

# Imprimir las etiquetas predichas para el conjunto de entrenamiento
print(y_pred_train)

# Calcular y mostrar la precisión del modelo en el conjunto de entrenamiento y prueba
print('Training set score: {:.4f}'.format(logreg.score(X_train, y_train)))
print('Test set score: {:.4f}'.format(logreg.score(X_test, y_test)))

from sklearn.metrics import confusion_matrix

# Suponiendo que tienes predicciones (y_pred) y etiquetas verdaderas (y_true)


# Calcular la matriz de confusión para el conjunto de prueba
conf_matrix = confusion_matrix(y_test, y_pred_test)

# Imprimir la matriz de confusión
print("Matriz de confusión:")
print(conf_matrix)

```
