```python
import requests
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import OneHotEncoder
```

```python
def query_data_paginated(url, params={}, headers={}, page_size=10000):
    """
    Queries data from the specified URL using paginated requests with offset and limit parameters.

    Args:
        url (str): The URL of the API endpoint.
        params (dict, optional): Additional query parameters to include. Defaults to {}.
        headers (dict, optional): HTTP headers to include in the request. Defaults to {}.
        page_size (int, optional): The number of records to retrieve per page. Defaults to 1000.

    Returns:
        list: A list of all retrieved data records.
    """

    all_data = []
    offset = 0

    while True:
        limit = page_size
        params["$offset"] = offset
        params["$limit"] = limit

        try:
            response = requests.get(url, headers=headers, params=params)
            response.raise_for_status()  # Raise an exception if request fails
        except requests.exceptions.RequestException as e:
            print(f"Error making request: {e}")
            break

        data = response.json() 

        if not data:  
            break

        all_data.extend(data)
        offset += limit

    return all_data

url = "https://www.datos.gov.co/resource/qhpu-8ixx.json"
headers = {"Accept": "application/json"} 

data = query_data_paginated(url, headers=headers)
df = pd.DataFrame(data)

print(df)
```

<h1>Hipótesis nula y alternativa para la predicción del valor de los fondos de inversión en Colombia</h1>
<h2>Hipótesis nula (H0):</h2>

No existe una relación significativa entre las variables independientes y el valor de los fondos de inversión en Colombia.

<h2>Hipótesis alternativa (H1)</h2>:

Sí existe una relación significativa entre las variables independientes y el valor de los fondos de inversión en Colombia.

<h2>Explicación:</h2>

La hipótesis nula (H0) establece que no hay ninguna relación entre las variables independientes (como el tipo de fondo, el principal compartimento, la rentabilidad histórica, etc.) y el valor de los fondos de inversión. En otras palabras, el valor de los fondos de inversión es completamente aleatorio y no se puede predecir con base en las variables independientes.
La hipótesis alternativa (H1) establece que sí existe una relación significativa entre las variables independientes y el valor de los fondos de inversión. Esto significa que el valor de los fondos de inversión se puede predecir, al menos en parte, con base en las variables independientes.
Verificabilidad:

Ambas hipótesis son verificables a través de métodos estadísticos. Para probar la hipótesis nula, se puede realizar una prueba de hipótesis. Si la prueba de hipótesis produce un valor p menor que el nivel de significancia elegido, entonces se rechaza la hipótesis nula y se acepta la hipótesis alternativa.

<h2>Mutua exclusión:</h2>

Las hipótesis nula y alternativa son mutuamente excluyentes. No pueden ser ambas verdaderas al mismo tiempo. Si la hipótesis nula es verdadera, entonces la hipótesis alternativa es falsa. Y viceversa.

<h2>Colectiva exhaustividad:</h2>

Las hipótesis nula y alternativa son colectivamente exhaustivas. Juntas, cubren todas las posibles situaciones en el contexto de predecir el valor de los fondos de inversión en Colombia. Si la hipótesis nula es falsa, entonces la hipótesis alternativa debe ser verdadera.

<h2>Ejemplo de aplicación:</h2>

Supongamos que queremos saber si existe una relación entre el principal compartimento de un fondo de inversión y su valor. La hipótesis nula sería que no existe ninguna relación, mientras que la hipótesis alternativa sería que sí existe una relación. Para probar estas hipótesis, se podría realizar una prueba de regresión. Si la prueba de regresión produce un valor p menor que el nivel de significancia elegido, entonces se rechazaría la hipótesis nula y se aceptaría la hipótesis alternativa. Esto significaría que existe una relación significativa entre el principal compartimento de un fondo de inversión y su valor.

<b>Contraste de hipótesis utilizando la biblioteca scipy.stats
A continuación, se presenta un ejemplo de cómo utilizar la biblioteca scipy.stats para llevar a cabo el contraste de hipótesis para la predicción del valor de los fondos de inversión en Colombia.</b>

```python
import numpy as np
from scipy import stats
```

```python
variable_dependiente = df["valor_fondo_cierre_dia_t"]
variables_independientes = df[["principal_compartimento", "rentabilidad_anual"]]
variables_independientes = pd.get_dummies(variables_independientes)
```

<b>En este caso, la prueba unilateral nos permite concluir que existe una relación significativa y positiva entre el principal compartimento y el valor del fondo de inversión.</b>

Configuración específica de la prueba:

Nivel de significancia: Se utiliza un nivel de significancia de 0.05, lo que significa que hay un 5% de probabilidad de rechazar la hipótesis nula cuando es verdadera.

```python
t_stat, p_valor = stats.ttest_ind(variable_dependiente, variables_independientes["principal_compartimento"])

if p_valor < 0.05:
    print("Existe una relación significativa y positiva entre el principal compartimento y el valor del fondo de inversión.")
else:
    print("No existe una relación significativa entre el principal compartimento y el valor del fondo de inversión.")
```

 <h2>Conclusión:</h2>

En este caso, podemos rechazar la hipótesis nula y concluir que existe una relación significativa entre el principal compartimento y el valor del fondo de inversión en Colombia.

<h2>Conclusión del contraste de hipótesis</h2>

El coeficiente de regresión positivo para la rentabilidad anual indica que, a mayor rentabilidad anual, mayor será el valor del fondo de inversión.
